```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load cleaned_orders.csv
df = pd.read_csv("ecommerce-orders.csv")

print("print First 5 Rows")
print(df.head())
```

```python
# Step 2: Basic Metrics
# Create TotalPrice column first
df['TotalPrice'] = df['Price'] * df['Quantity']

total_orders = df.shape[0]
total_revenue = df['TotalPrice'].sum()
avg_order_value = df['TotalPrice'].mean()

print("Total Orders:", total_orders)
print("Total Revenue:", total_revenue)
print("Average Order Value:", avg_order_value)
```

```python

# Step 3: Top 3 Cities by total sales
top_cities = df.groupby('City')['TotalPrice'].sum().sort_values(ascending=False).head(3)
print(top_cities)
```

```python

# Step 4: Most popular category by quantity sold
popular_category = df.groupby('Category')['Quantity'].sum().sort_values(ascending=False).head(1)
print(popular_category)
```

```python
# step 5 : group by gender
print(df.columns)
```

```python
df['TotalPrice'] = df['Price'] * df['Quantity']
```

```python
gender_group = df.groupby('Gender').agg(
    Average_Order_Value = ('TotalPrice', 'mean'),
    Total_Sales = ('TotalPrice', 'sum')
)

print(gender_group)
```

```python
# Step 6: Visualizations

plt.figure()   # Create a new blank figure (a canvas) for plotting graphs.
df.groupby('City')['TotalPrice'].sum().plot(kind='bar')
plt.title("Total Sales by City")
plt.ylabel("Sales")
plt.show()

plt.figure()
df.groupby('Gender')['TotalPrice'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Gender")
plt.ylabel("")
plt.show()
```

```python
# Step 7: Insights Summary
print("1. A few cities contribute majority of total revenue.")
print("2. One product category dominates in quantity sold.")
print("3. Spending pattern varies between genders in both AOV and total sales.")
```
